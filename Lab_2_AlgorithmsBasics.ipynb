```python
# Exercise 4
def func4(a, b):
    b.sort() # Time complexity for this function is O(Nlog N) where N = len(b)
    intersect = 0
    for x in a:
        if binary_search(b, x) >= 0:
            intersect = intersect + 1
    return intersect

def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

```python
# Exercise 5
def func5(n):
    guess = 1
    while guess * guess <= n:
        if guess * guess == n:
            return guess
        guess = guess + 1
    return None
```

```python
# Sample function call: func5(16)
import time

start_time=time.time()
x = func5(16)
print(x)
end_time=time.time()

print((end_time-start_time)*1000)
```

```python
# Exercise 6
def func8(A):
    func8_helper(A, 0, len(A) -1)

def func8_helper(A, i, j):
    #Input: An array A and two indices i and j.
    n = j - 1 + 1
    if n ==2:
        if A[i] > A[j]:
            # Swap A[i] and A[j]
            A[i], A[j] = A[j], A[i]
        elif n > 2:
            m = n // 3
            func8_helper(A, i, j - m)
            func8_helper(A, i + m, j)
            func8_helper(A, i, j - m)
        return A
```

```python
# Exercise 7
def func12(arr, k):
    arr_len = len(arr)
    k = k % arr_len
    # Initialize a new array of the same length as 'arr'
    custom_arr = [0] * arr_len

    for i in range(arr_len):
        mystery_index = (i + k) % arr_len
        custom_arr[mystery_index] = arr[i]
    return custom_arr

```

```python
#write a function named as max_min to find minimum (or maximum) element in an array
def max_min(arr, find="max"):
    if not arr:  # handle empty array
        return None

    if find == "max":
        return max(arr)
    elif find == "min":
        return min(arr)
    else:
        raise ValueError("find must be 'max' or 'min'")


arr = [12, 1234, 45, 67, 1]
print("Maximum:", max_min(arr, "max"))  # Output: 1234
print("Minimum:", max_min(arr, "min"))  # Output: 1

```

```python
# write a function named as Counting_EvenOdd to count number of even and odd elements in an array
def Counting_EvenOdd(arr):

    count_even = 0
    count_odd = 0

    for num in arr:
        if num % 2 == 0:
            count_even += 1
        else:
            count_odd += 1

    return count_even, count_odd



arr = [12, 1234, 45, 67, 1]
even_count, odd_count = Counting_EvenOdd(arr)
print("Even elements:", even_count)  # Output: 2
print("Odd elements:", odd_count)    # Output: 3
```

```python
# write a function named as Binary_Search to perform binary search in 1D-array
def Binary_Search(arr, target):

    left = 0
    right = len(arr) - 1

    while left <= right:
        mid = (left + right) // 2  # middle index

        if arr[mid] == target:
            return mid  # target found
        elif arr[mid] < target:
            left = mid + 1  # search right half
        else:
            right = mid - 1  # search left half

    return -1  # target not found



arr = [1, 3, 5, 7, 9]
print("Index of 5:", Binary_Search(arr, 5))   # Output: 2
print("Index of 9:", Binary_Search(arr, 9))   # Output: 4
print("Index of 2:", Binary_Search(arr, 2))   # Output: -1
```
